##Linear Regression

House Pricing Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

filename = 'housing.csv'
df = pd.read_csv(filename)
df

df.rename(columns={'MEDV':'Price'}, inplace=True)
df
```

```python
dataset = pd.DataFrame(df)
print(dataset.head(6))
```

```python
dataset.head(10)
```

Dividing the dataset into dependent and independent features

```python
x = dataset.iloc[:,:-1]  ##specifies all the columns except the last one , first : represents all the data , then :-1 represent all the data from start to last-1
y = dataset.iloc[:,-1]  ##specifies the last column
```

```python
###LINEAR REGRESSION

from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import cross_val_score

#cross validation : In cross-validation, the dataset is divided into multiple subsets. 
# The model is trained on some of these subsets and tested on others, iteratively. 
# This helps evaluate the model's performance more reliably than a single train-test split, 
# ensuring it generalizes well to unseen data. Common types include k-fold cross-validation and leave-one-out cross-validation.
linear = LinearRegression() ##creating isntance of the model
mse = cross_val_score(linear,x,y,scoring = 'neg_mean_squared_error',cv=5) #first parameter is my model
#cv is cross validation value
#we can alsp takle multiple cv, just add a for loop.
# MSE  = mean squared error / also known as cost function
print(mse)
mean_mse = np.mean(mse)
print("Mean of MSE:", mean_mse)
```

```python
##Ridge Regression , in order to overcome overfitting and feature selection we use ridge and laso regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

#GridSearchCV systematically explores a predefined hyperparameter grid for a machine learning model, training and evaluating the model with different combinations using cross-validation. It helps find the optimal hyperparameters that lead to the best model performance. The final model is then trained with the identified best hyperparameters.

ridge = Ridge()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regresion = GridSearchCV(ridge,parameters,scoring="neg_mean_squared_error",cv=5)
ridge_regresion.fit(x,y)
```

```python
print(ridge_regresion.best_params_)
print(ridge_regresion.best_score_)
```

```python
## Here mse increased , so ridge regression isnt helping us out.
##because previous we got -37 (linear ) and in ridge we got -32. 
#So in comparison to both linear is more useful

# laso Regressio n

from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()

parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-1,1,5,10,20]}

lasso_regression = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(x,y)
```

```python

print(lasso_regression.best_params_)
print(lasso_regression.best_score_)
```

```python
# so while comparing all the regression values
# linear = -37
#ridge = -29
# lasso = - 35

## in ideal case the value should approach to zero
```

```python
#Now lets increase alpha value and check working of lasso and ridge

params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-1,1,5,10,20,30,40,50,60,70,100,110]}
lasso_regression = GridSearchCV(lasso,params,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(x,y)

ridge_regresion = GridSearchCV(ridge,params,scoring="neg_mean_squared_error",cv=5)
ridge_regresion.fit(x,y)
```

```python
print("Lasso Regression best parameters are : ",lasso_regression.best_params_)
print("Lasso Regression best score is : ",lasso_regression.best_score_)

print("Ridge Regression best parameters are : ",ridge_regresion.best_params_)
print("Ridge Regression best score is : ",ridge_regresion.best_score_)

```

_____ TRAIN AND TEST SPLIT ___


```python
##TRAIN AND TEST SPLIT
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 0)
```

```python
# FOR LINEAR REGRESSION

###LINEAR REGRESSION

from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import cross_val_score

linear = LinearRegression()
linear.fit(x_train,y_train)
mse = cross_val_score(linear,x_train,y_train,scoring = 'neg_mean_squared_error',cv=5) #first parameter is my model
print(mse)
mean_mse = np.mean(mse)
print("Mean of MSE:", mean_mse)

```

#### if value nearer to zero , Better the performance 

```python
#for RIDGE REG

from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV


ridge = Ridge()
parameters = {'alpha':[1e-15,1e-10,1e-8,1e-2,1,5,10,20]}
ridge_regresion = GridSearchCV(ridge,parameters,scoring="neg_mean_squared_error",cv=5)
ridge_regresion.fit(x_train,y_train)

print(ridge_regresion.best_params_)
print(ridge_regresion.best_score_)

#performed better thn linear , as its more clsoer to zero
```

```python
# FOR LASSO reg
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso = Lasso()

parameters = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1e-1,1,5,10,20]}

lasso_regression = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regression.fit(x_train,y_train)

print(lasso_regression.best_params_)
print(lasso_regression.best_score_)

##performed slightly bad than ridge but better than linear
```

```python
#predicting values using test cases for lasso

yPredict = lasso_regression.predict(x_test)

#we will be suing r square and r square adjusted
from sklearn.metrics import r2_score

r2_score_1 = r2_score(yPredict,y_test)
print(r2_score_1)

#should ne nearer to 1 (i.e 100%)
```

```python
#predicting values using test cases for Ridge

yPredict = ridge_regresion.predict(x_test)

#we will be suing r square and r square adjusted
from sklearn.metrics import r2_score

r2_score_1 = r2_score(yPredict,y_test)
print(r2_score_1)

#should ne nearer to 1 (i.e 100%)
```

```python
#predicting values using test casesfor linear Regression

yPredict = linear.predict(x_test)

#we will be suing r square and r square adjusted
from sklearn.metrics import r2_score

r2_score_1 = r2_score(yPredict,y_test)
print(r2_score_1)

#should ne nearer to 1 (i.e 100%)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
#for logistic we need classification dataset
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
df
```

```python
##Independent Features
x = pd.DataFrame(df['data'],columns=df['feature_names'])
x.head(5)
```

```python
## Dependent Features
#3 all depenendent features present in target
y = pd.DataFrame(df['target'],columns=['Target'])
y

#0 means , no breast cancer
#1 means, yes there is breast cancer
```

```python
##checking if the dataset is Balanced or not
# if dataset imbalanced we do up sampling
y['Target'].value_counts()
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
## logistic regression , we have l1 and l2 regularization/norms , we also have C (1 / lambda), class_weights(used to apply on our data set when it's imbalanced)
param = [{'C':[1,5,10]},{'max_iter':[100,150]}] #max iter used for gridsearchCV
model1 = LogisticRegression(C=100,max_iter=100)
model1.fit(x_train,y_train)
```

```python
model = GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
#Since its classification problem , and we ain't sure about 
# True positive/negative we will use f1 scoring

model.fit(x_train,y_train)
```

```python
print(model.best_params_)
print(model.best_score_)
```

```python
#predicting on test data

y_pred= model.predict(x_test)
y_pred
```

```python
## now we will be applying confusion matrix
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
## accuracy test using 
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
