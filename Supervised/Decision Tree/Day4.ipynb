```python
import pandas as pandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
## we have 4 independent features here
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier() 
x = iris.data
y = iris.target
model.fit(x,y)
##right now we will be doing overfitting
```

```python
##we have hyper parameters they are criteria ginni , then splitter , max_depth
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
tree.plot_tree(model,filled = True)
```
