```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
##load The cancer
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
cancer
```

```python
print(cancer.DESCR)
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df
```

```python
#Perform Dimensionality reduction
#perform feature Extraction using PCA

#apply standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#storing scaler value in df
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
scaled_data

# here standardization applied , i.e mean=0
# std dev = 1 , 
#most of data ranges -3  to + 3
```

```python
# applying PCA algorithm

from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
#n_components = 2 , means convert all to 2 features, capturing max variance
```

```python
data_pca = pca.fit_transform(scaled_data )
print(data_pca)

#it will have just 2 columns, so out of all the columns PCA have been ap;lied and transformation have been aplied
#eigen decomposition is used.
#thus this is called dimensionality reduction
```

```python
pca.explained_variance_
#shows variance of feature 1 and 2 captured
```

```python
#visualize PCA
plt.figure(figsize=(10,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer['target'],cmap='plasma')
#data_pca[:,0] means 0th feature of our data_pca
#always for  variance:  0th feature  > 1st feature
#data_pca[:,1] means 1st feature of our data_pca
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
print("Explained variance ratio (PCA)")
print(pca.explained_variance_ratio_)
print("Explained variance (PCA)")
print(pca.explained_variance_)

```

```python
# for 3 features
pca = PCA(n_components=3)
data_pca = pca.fit_transform(scaled_data )
print(data_pca)
```

```python
#visualize the PCA
plt.figure(figsize=(10,6))
plt.scatter(data_pca[:,0],data_pca[:,1],data_pca[:,2],c=cancer['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
print("Explained variance ratio (PCA)")
print(pca.explained_variance_ratio_)
print("Explained variance (PCA)")
print(pca.explained_variance_)
#here we can see that 3rd feature is not that important
#so we can drop it
```
